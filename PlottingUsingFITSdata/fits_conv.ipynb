```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.stats import sigma_clipped_stats

%matplotlib inline

file_path = 'ngc1087/hlsp_phangs-hst_hst_wfc3-uvis_ngc1087_f275w_v1_exp-drc-sci.fits'
```

```python
img=fits.open(file_path)
img.info()
```

```python
fits_data=img[0].data[1000:7000,1000:7000]
print(fits_data.shape, fits_data.dtype)
```

```python
mean, median, std = sigma_clipped_stats(fits_data)
```

```python
plt.figure(figsize=(10,10))

plt.imshow(fits_data, vmin = median - 3*std, vmax = median + 3*std, cmap='gray')
plt.title('Image of NGC 1087 at 275 nm')
plt.axis('equal')
plt.axis('off')
plt.show()
```
